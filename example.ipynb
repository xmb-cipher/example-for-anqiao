```python
# pandas and numpy are not python built-in
# make sure you have them installed
import pandas as pd
import numpy as np
import glob
```

```python
# relace this with your files
fnames = glob.glob('/Users/mingbinxu/Desktop/anqiao-*.csv')
```

```python
dtype = {'凭证类型': str, '总账': str, '本位币金额': np.float32}
df = pd.concat([pd.read_csv(fname, header='infer', dtype=dtype) for fname in fnames])

# show the first few rows
df.head()
```

```python
result = df[dtype.keys()].groupby(['凭证类型', '总账']).sum('本位币金额')
```
